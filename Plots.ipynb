```python
import matplotlib.pyplot as plt

# --- 1. Data Setup (Uniform 30 Epochs) ---
epochs = [1, 5, 10, 15, 20, 25, 30]

# Model 1: Std Transformer
val_m1 = [6.6718, 5.1427, 4.6801, 4.5405, 4.4976, 4.4842, 4.4804]
acc_m1 = [9.4, 18.6, 22.9, 24.9, 25.8, 26.2, 26.3]
ppl_m1 = [789.8, 171.2, 107.8, 93.7, 89.8, 88.6, 88.3]

# Model 2: Wave Field LLM (Original)
val_m2 = [7.2050, 6.2668, 5.6538, 5.3991, 5.2187, 5.1467, 5.1276]
acc_m2 = [4.6, 10.4, 14.2, 15.9, 17.5, 18.4, 18.7]
ppl_m2 = [1346.1, 526.8, 285.4, 221.2, 184.7, 171.9, 168.6]

# Model 3: Wave Field LLM Modified (Truncated to 30 Ep)
val_m3 = [7.2044, 6.4286, 5.7035, 5.3068, 5.0745, 4.9571, 4.8912]
acc_m3 = [4.6, 8.9, 13.7, 16.6, 19.2, 20.8, 21.7]
ppl_m3 = [1345.3, 619.3, 299.9, 201.7, 159.9, 142.2, 133.1]

# --- 2. Create the Layout ---
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Performance Comparison: Standard vs Wave Field Architectures (30 Epochs)', fontsize=16)

# Graph 1: Validation Loss
ax1.plot(epochs, val_m1, 'o-', label='Model 1: Std Transformer', color='tab:blue', linewidth=2)
ax1.plot(epochs, val_m2, 's-', label='Model 2: Wave Field LLM (Original)', color='tab:red', linewidth=2)
ax1.plot(epochs, val_m3, '^-', label='Model 3: Wave Field LLM Modified', color='tab:green', linewidth=2)
ax1.set_title('Validation Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Graph 2: Accuracy
ax2.plot(epochs, acc_m1, 'o-', label='Model 1: Std Transformer', color='tab:blue', linewidth=2)
ax2.plot(epochs, acc_m2, 's-', label='Model 2: Wave Field LLM (Original)', color='tab:red', linewidth=2)
ax2.plot(epochs, acc_m3, '^-', label='Model 3: Wave Field LLM Modified', color='tab:green', linewidth=2)
ax2.set_title('Validation Accuracy %')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Graph 3: Perplexity (PPL)
ax3.plot(epochs, ppl_m1, 'o-', label='Model 1: Std Transformer', color='tab:blue', linewidth=2)
ax3.plot(epochs, ppl_m2, 's-', label='Model 2: Wave Field LLM (Original)', color='tab:red', linewidth=2)
ax3.plot(epochs, ppl_m3, '^-', label='Model 3: Wave Field LLM Modified', color='tab:green', linewidth=2)
ax3.set_title('Perplexity (PPL)')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('PPL')
ax3.set_yscale('log') 
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
